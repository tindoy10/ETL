```python
import pandas as pd
```

```python
# Extract: Load the dataset
file_path = 'train.csv'
data = pd.read_csv(file_path)

display(data)
```

```python
# Transform: Data Cleaning and Transformation

# 1. Handle missing values by dropping rows with missing Postal Codes
data_cleaned = data.dropna(subset=['Postal Code'])

# 2. Convert 'Order Date' and 'Ship Date' to datetime format (original format: day/month/year)
data_cleaned['Order Date'] = pd.to_datetime(data_cleaned['Order Date'], format='%d/%m/%Y')
data_cleaned['Ship Date'] = pd.to_datetime(data_cleaned['Ship Date'], format='%d/%m/%Y')

# 3. Reformat the 'Order Date' and 'Ship Date' to month/day/year
data_cleaned['Order Date'] = pd.to_datetime(data_cleaned['Order Date']).dt.strftime('%m/%d/%Y')
data_cleaned['Ship Date'] = pd.to_datetime(data_cleaned['Ship Date']).dt.strftime('%m/%d/%Y')

# 4. Remove duplicates if any
data_cleaned = data_cleaned.drop_duplicates(keep=False)

# 5. Handle outliers in the 'Sales' column using IQR (Interquartile Range)
Q1 = data_cleaned['Sales'].quantile(0.25)  # 25th percentile = 17.248
Q3 = data_cleaned['Sales'].quantile(0.75)  # 75th percentile = 210.392
IQR = Q3 - Q1  # Interquartile Range = 193.144
```

```python
print(f"25th percentile (Q1): {Q1}")

print(f"75th percentile (Q3): {Q3}")
```

```python
# Define lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR # Lower bound: 0
upper_bound = Q3 + 1.5 * IQR # Upper bound: 500.108

print(f"Lower bound: {lower_bound}")

print(f"Upper bound: {upper_bound}")

# Filter out outliers
data_cleaned = data_cleaned[(data_cleaned['Sales'] >= lower_bound) & (data_cleaned['Sales'] <= upper_bound)]
```

```python
# Load: Save the cleaned data to a new CSV file
output_file_path = 'cleaned_train.csv'
data_cleaned.to_csv(output_file_path, index=False)

print(f"Data has been cleaned and saved to {output_file_path}")
```
